```python
#Normalization
#To have a mean 0 and standard devaition 1

data=np.array([1,2,3,4,5])

#Mean and Standard deviation
mean=np.mean(data)
standard_devn=np.std(data)


#Normalize the data
normalized_data=(data-mean)/standard_devn

print("Normalized data :", normalized_data)

var=np.var(data)
print(var)
```

```python
import pandas as pd

de=pd.read_csv('data.csv')
de.head(5)
de.tail(5)
de.describe()
de.dtypes
```

```python
#Handling missing values
# de[de.isnull()]
# de.isnull().any()
de.isnull().sum()
```

```python
de.fillna(2)
```

```python
de['Duration_fillNA']=de['Duration'].fillna(de['Duration'].mean())
de
```

```python
# de.dtypes
#Renaming columns

de=de.rename(columns={'Date':'Diet Date'})
de.head(5)

#Changing Datatype
de['Pulse_new']=de['Pulse'].astype(float)
de['calories_new']=de['Calories'].fillna(de['Calories'].mean()).astype(int)
de.head(5)
```

```python
# de['New Pulse']=de['Pulse'].apply(lambda x:x**2)
# de

#Data aggregating and grouping

# grp_mean=de.groupby('Calories')['Pulse'].mean()
# print(grp_mean)

grp_sum=de.groupby(['Duration'],['maxpulse'])['Pulse'].sum()
property(grp_sum)
```

```python
grp_agg=de.groupby('Pulse')['Maxpulse'].agg(['mean','sum','count'])
grp_agg
```

```python
#MErging and Joining
df1=pd.DataFrame({'Key':['A','B','C'],'Value1':[1,2,3]})
df2=pd.DataFrame({'Key':['A','B','D'],'Value1':[4,5,6]})

pd.merge(df1,df2,on="Key",how='outer')
```
