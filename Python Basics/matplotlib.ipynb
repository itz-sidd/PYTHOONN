```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[1,4,9,16,25]

#Creating a line plot
plt.plot(x,y)
plt.xlabel('X asix')
plt.ylabel('Y label')
plt.title('Basic line plot')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y,color='green',linestyle=':',marker='o',linewidth=3,markersize=10)
plt.grid(True)
```

```python
#Multiple Plot
#Sample Data

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green',marker='*',markersize=9)
plt.title('Plot 1')
plt.grid(True)

plt.subplot(2,2,2)
plt.plot(y1,x,color='blue',marker='o',markersize=9)
plt.title('Plot 2')

plt.subplot(2,2,3)
plt.plot(x,y1,color='yellow',marker='*',markersize=9)
plt.title('Plot 3')

# z=x**2

plt.subplot(2,2,4)
plt.plot(y2,x,color='green',marker='*',markersize=9)
plt.title('Plot 4')
```

```python
#Bar Plot

categories=['A','B','C','D','E']
values=[5,6,17,8,12]

plt.bar(categories,values,color='purple')
```

```python
#Histograms

data=[1,2,3,4,5,6,7,8,2,2,5,4,4,5,6]

plt.hist(data,bins=20,color='orange',edgecolor='black')
```

```python
x=[1,2,3,4,5]
y=[2,3,4,5,6]

plt.scatter(x,y,color='red',marker='*')
```

```python
#Pie chart

labels=['A','B','C','D']
sizes=[30,40,50,20]
colors=['green','gold','orange','red']
explode=(0.2,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%",shadow=True)
```

```python
#Sales data Visualisation

import pandas as pd

sales_data_df = pd.read_csv('sales_data_sample.csv', encoding='latin1')
sales_data_df.head(5)
```

```python
sales_data_df.info()
```

```python
#Plotting total sales by price each

total_sales_by_product=sales_data_df.groupby('PRICEEACH')['SALES'].sum()
print(total_sales_by_product)
```

```python
total_sales_by_product.plot(kind='bar',color='teal')
```

```python
sales_trend=sales_data_df.groupby('SALES')['QUANTITYORDERED'].sum().reset_index()
sales_trend

plt.plot(sales_trend['QUANTITYORDERED'],sales_trend['SALES'])


```

```python
import seaborn as sns

#Basic plotting

tips=sns.load_dataset('tips')
tips
```

```python
#Creating a scatter plot

import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title('Scatter plot of total bills')
plt.show()
```

```python
##Line plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title('Line plot of total size/total bill')
plt.show()
```

```python
#Categorical Plots

sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar plot of total bill/day')
plt.show()
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
plt.show()

sns.violinplot(x='day',y='tip',data=tips)
plt.show()
```

```python
#HIstograms

sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
#KDE plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
sns.pairplot(tips)
```

```python
#Heatmap
correlation=tips[['total_bill','tip','size']].corr()
correlation
```

```python
import seaborn as sns
correlation = tips.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
